```python
import numpy as np
```

```python
#collecting data 
# rows = student [AKASH,RIYA,KAJAL,PRIYA,RAHUL,MAYANK].
# column = subject[MATHS,SCIENCE,ENGLISH,HINDI,SOCIAL SCIENCE] respectively.
```

```python
marks = np.array((
        [90,87,67,46,87],
        [98,78,33,87,88],
        [78,92,44,23,93],
        [29,99,66,59,80],
        [92,87,93,95,86],
        [39,94,63,45,82]  
))

students =["Akash","Riya","Kajal",'Priya','Rahul','Mayank']
subjects = ['Maths','Science','English','Hindi','Social science']
print(students)
print(subjects)
print(marks)
```

Student-wise averge helps to evaluate individual performance.(row wise)

```python
student_average = np.mean(marks,axis=1)
for i in range(len(students)) :
       print(students[i],"=" ,student_average[i])
```

Subject-wise average helps to evaluate overall class strength and weekness

```python
subject_average = np.mean(marks,axis=0)
for i in range(len(subjects)):
      print(subjects[i],"=",subject_average[i])
```

max and min marks of student in subjects

```python
marks
```

```python
max_marks = np.max(marks,axis=1)
for i in range(len(students)):
         print(students[i],"=",max_marks[i])
```

```python
min_marks = np.min(marks,axis=1)
for i in range(len(students)):
    print(students[i],"=",min_marks[i])
```

max and min marks of subject each student

```python
max_marks_subject = np.max(marks,axis=0)
for i in range(len(subjects)):
         print(subjects[i],"=",max_marks_subject[i])
```

```python
min_marks_subject = np.min(marks, axis=0)
for i in range(len(subjects)):
    print(subjects[i],"=", min_marks_subject[i])
```
